```python
# Importing packages and libraries
import os
import cv2
import time
import matplotlib.pyplot as plt 
from skimage.filters import threshold_otsu
import numpy as np
from glob import glob
from scipy import misc
from matplotlib.patches import Circle, Ellipse
from matplotlib.patches import Rectangle
from PIL import Image
from sklearn.model_selection import train_test_split

import gzip
from keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D
from keras.models import Model
from keras.optimizers import RMSprop
from keras.layers.normalization import BatchNormalization

# Setting environment variables to use the First GPU
os.environ["CUDA_DEVICE_ORDER"] = "PCI_BUS_ID"
os.environ["CUDA_VISIBLE_DEVICES"] = "0"

# Magic Command for viewing Matplotlib plots
%matplotlib inline
```

```python
# Reading File Names and Printing No.of Files
data = glob(".data/DB*/*")
print("No.of Images - {}".format(len(data)))
```

```python
# Function to read and resize the images
images = []
def read_images(data):
    for i in range(len(data)):
        img = misc.imread(data[i])
        img = misc.imresize(img, (224, 224))
        images.append(img)
    return images
```

```python
# Reading Images and Storing them as a tensor
images = read_images(data)
images_arr = np.asarray(images)
images_arr = images_arr.astype('float32')
```

```python
# Printing shape of Images
print("Dataset (images) Shape : {shape}".format(shape = images_arr.shape))
```

```python
# Displaying the first two images
for i in range(2):
    plt.figure(figsize = (5, 5))
    curr_img = np.reshape(images_arr[i], (224, 224))
    plt.imshow(curr_img, cmap = 'gray')
    plt.show()
```

```python
# Resizing Image Array to required keras shape 
images_arr = images_arr.reshape(-1, 224, 224, 1)

# Normalising Images Dividing by Maximum Pixel Value
images_arr = images_arr / np.max(images_arr)
```

```python
# Splitting into train and validation sets
train_X, valid_X, train_ground, valid_ground = train_test_split(images_arr, images_arr, test_size = 0.2, 
                                                                random_state =13)
```

```python
# Setting the hyperparameters
batch_size = 8
epochs = 200
inChannel = 1
x, y = 224, 224
input_img = Input(shape = (x, y, inChannel))
```

```python
# Function to create the autoencoder architecture
def make_autoencoder(input_img):
    
    # Encoder
    # input = 28 x 28 x 1(wide and thin)
    conv1 = Conv2D(32, (3, 3), activation = 'relu', padding = 'same')(input_img) # 28 x 28 x 32 
    pool1 = MaxPooling2D(pool_size = (2, 2))(conv1) # 14 x 14 x 32 
    conv2 = Conv2D(64, (3, 3), activation = 'relu', padding = 'same')(pool1) # 14 x 14 x 64
    pool2 = MaxPooling2D(pool_size = (2, 2))(conv2) # 7 x 7 x 64
    conv3 = Conv2D(128, (3, 3), activation = 'relu', padding = 'same')(pool2) # 7 x 7 x 128(small and thick)
    
    # Decoder
    conv4 = Conv2D(128, (3, 3), activation = 'relu', padding = 'same')(conv3) # 7 x 7 x 128(small and thick)
    up1 = UpSampling2D((2, 2))(conv4)  # 14 x 14 x 128
    conv5 = Conv2D(64, (3, 3), activation = 'relu', padding = 'same')(up1) # 14 x 14 x 64
    up2 = UpSampling2D((2, 2))(conv5) # 28 x 28 x 64
    decoded = Conv2D(1, (3, 3), activation = 'sigmoid', padding = 'same')(up2) # 28 x 28 x 1
    return decoded
```

```python
# Making and compiling the autoencoder with loss = 'mean_squared_error'
# and optimizer RMSprop
autoencoder = Model(input_img, make_autoencoder(input_img))
autoencoder.compile(loss='mean_squared_error', optimizer = RMSprop())
```

```python
# Visualising the autoencoder model
autoencoder.summary()
```

```python
# Training the autoencoder
start = time.time()
autoencoder_train = autoencoder.fit(train_X, train_ground, batch_size = batch_size, 
                                    epochs = epochs, verbose = 1, 
                                    validation_data = (valid_X, valid_ground))
end = time.time()
```

```python
# Displaying total time taken to train model
print("Time Taken to train model - {}s".format(end - start))
```

```python
# Plot validation and training loss with respect to number of epochs
loss = autoencoder_train.history["loss"]
val_loss = autoencoder_train.history['val_loss']
epochs = range(200)
plt.figure()
plt.plot(epochs, loss, label = 'Training Loss', color = 'b')
plt.plot(epochs, val_loss, label = 'Validation Loss', color = 'r')
plt.title('Training and Validation Loss')
plt.legend()
plt.show()
```

```python
# Save the autoencoder model weights to disc
autoencoder = autoencoder.save_weights('./model/autoencoder.h5')
```

```python
# Loading model weights from disc and compiling for prediction
autoencoder = Model(input_img, make_autoencoder(input_img))
autoencoder.load_weights('./model/autoencoder.h5')
autoencoder.compile(loss='mean_squared_error', optimizer = RMSprop())
```

```python
# Reconstructing Validation Set Images  
pred = autoencoder.predict(valid_X)
```

```python
# Display Test Images and Reconstructed Images
plt.figure(figsize = (20, 4))
print("Test Images")
for i in range(5):
    plt.subplot(1, 5, i+1)
    plt.imshow(valid_ground[i, ..., 0], cmap = 'gray')
plt.show()

plt.figure(figsize = (20, 4))
print("Reconstruction of Test Images")
for i in range(5):
    plt.subplot(1, 5, i+1)
    plt.imshow(pred[i, ..., 0], cmap = 'gray')
plt.show()
```
